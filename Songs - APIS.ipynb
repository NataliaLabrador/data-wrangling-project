# Add new variables 

## Column album

```python
# Create a function to obtain the album of each song based on its id
def get_album(track_id):
    try:
        # Obtenain information of each song
        track_info = sp.track(track_id)
        # Get album name
        album_name = track_info['album']['name']
        return album_name
    except:
        return None

# Create a column called album and apply the function to every row in the track_id column
songs_apis_df['album'] = songs_apis_df['track_id'].apply(get_album)
```

## Column duration

```python
# Create a function to obtain the duration of each song based on its id
def get_duration(track_id):
    try:
        # Obtain the information of each song
        track_info = sp.track(track_id)
        # Obtain the duration of each song
        duration_ms = track_info['duration_ms']
        # Convert the duration into seconds 
        duration_sec = duration_ms / 1000
        return duration_sec
    except:
        return None

# Create a column called duration and apply the function to every row in the track_id column
songs_apis_df['duration'] = songs_apis_df['track_id'].apply(get_duration)
```
