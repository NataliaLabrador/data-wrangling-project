# Conclusions based on variables

## Column released_year + released_month

### Range of release year and month

```python
# Find the earliest release date
min_date = songs_final_df[['released_year', 'released_month']].astype(str).agg('-'.join, axis=1).min()

# Find the latest release date
max_date = songs_final_df[['released_year', 'released_month']].astype(str).agg('-'.join, axis=1).max()

print("The date range goes from", min_date, "to", max_date)
```

### Percentage of songs released in 2023

```python
# Check how many songs were released in 2023 
songs_2023 = songs_final_df[songs_final_df['released_year'] == 2023].shape[0]

# Calculate total songs 
total_songs = songs_final_df.shape[0]

# Calculate percentage of songs released in 2023 
percentage_2023 = (songs_2023 / total_songs) * 100

print("The percentage of songs released in 2023 is:", percentage_2023, "%")
```

### Most common year

```python
# Count the frequency of each release year
years_frequency = songs_final_df['released_year'].value_counts()

# Get the most common year (highest frequency)
most_common_year = years_frequency.idxmax()

# Get the frequency of the most common year
most_common_frequency = years_frequency.max()

print("The most common year where songs were released is:", most_common_year)
print("Number of released songs in the most common year:", most_common_frequency)
```

### Percentage of songs released in 2022

```python
# Check how many songs were released in 2022 
songs_2022 = songs_final_df[songs_final_df['released_year'] == 2022].shape[0]

# Calculate total songs 
total_2022_songs = songs_final_df.shape[0]

# Calculate percentage of songs released in 2022 
percentage_2022 = (songs_2022 / total_songs) * 100

print("The percentage of songs released in 2022 is:", percentage_2022, "%")
```
