### Column streams:

```python
# Check for unique values in column streams
songs_df['streams'].unique()
```

```python
# Convert the column streams into numeric type
songs_df.loc[:, 'streams'] = pd.to_numeric(songs_df['streams'], errors='coerce')

# Delete rows that contain non numerical values in the streams column
songs_df = songs_df[songs_df['streams'].notna()]

# Convert the column streams into integer type
songs_df.loc[:, 'streams'] = songs_df['streams'].astype(int)
```

```python
# Sort by streams in ascending order
songs_df_sorted = songs_df.sort_values(by='streams', ascending=False)
```

```python
# Reduce the DataFrame to 200 
reduced_df = songs_df_sorted.head(500)
```

```python
# Reset the index
sorted_df = reduced_df.reset_index(drop=True)
sorted_df
```

```python
# Save the cleaned DataFrame in a pickle file
sorted_df.to_pickle('sorted_df.pkl')
```
