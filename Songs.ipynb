```python
# Get tracks ids
track_ids = []

# FunciÃ³n para buscar una canciÃ³n y obtener su track_id
#def get_track_id(song_title, artist):
    results = sp.search(q='track:{} artist:{}'.format(song_title, artist), type='track', limit=1)
    if results['tracks']['items']:
        return results['tracks']['items'][0]['id']
    else:
        return None

```

```python
# FunciÃ³n para obtener el ID de una canciÃ³n por su nombre
def get_track_id(track_name):
    try:
        # Realizar una bÃºsqueda de la canciÃ³n por su nombre
        results = sp.search(q=track_name, type='track', limit=1)
        # Extraer el ID de la primera canciÃ³n encontrada
        track_id = results['tracks']['items'][0]['id']
        return track_id
    except:
        # Manejar el caso en que no se encuentre la canciÃ³n
        return None

# Aplicar la funciÃ³n get_track_id a cada fila de la columna 'track_name'
songs_df['track_id'] = songs_df['track_name'].apply(get_track_id)
```

```python
# Aplicar la funciÃ³n get_track_id a cada fila del DataFrame
songs_df['track_id'] = songs_df.apply(lambda row: get_track_id(row['track_name'], row['artist(s)_name']), axis=1)
```

```python
# Agregar los track IDs al DataFrame
songs_df['track_id'] = track_ids
```

```python
# Create a new column with the gender of each song using its track_id
def get_genres(track_id):
    try:
        track_info = sp.track(track_id)
        return track_info['album']['genres']
    except:
        return None

# Apply the function to the column track_id to obtain the genres
songs_df['genres'] = songs_df['track_id'].apply(get_genres)
```
